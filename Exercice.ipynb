# 1️ Importer le jeu de données

```python
import pandas as pd
df = pd.read_csv('prets.csv')
df
```

## 2. Explorer la structure du dataFrame 

```python
print("les attributs:",df.columns)
print("les dimensions",df.shape)
print("nombre de villes uniques:",df['ville'].nunique())
```

## 3. Créer un nouvel attribut « taux_endettement »

```python
df['taux_endettement'] = round(df['remboursement'] * 100 / df['revenu'], 2)
df
```

# 4. Renommer une colonne

```python
df.rename(columns={'taux':'taux_interet'}, inplace=True)
df
```

# 5 créer autres 

```python
df['cout_total'] = df['remboursement'] * df['duree']
df['benefices'] = (df['cout_total'] * df['taux_interet']) / 24
df
```

```python
df.nlargest(5, 'benefices')
```

8. Ajouter une variable nommée « risque » qui a comme valeur « Oui » si le client est
risqué (taux d’endettement > 40%), « Non » sinon

```python
df['risque'] = df['taux_endettement'].apply(lambda x: 'Oui' if x > 40 else 'Non')
df
```

```python
df['risque'] = 'Non'  # Valeur par défaut
df.loc[df['taux_endettement'] > 40, 'risque'] = 'Oui'
```

9. Récupérer le nombre de prêts automobiles accordés ? Quel est le coût total moyen de
ces derniers ? 

```python
prets_auto = df[df['type'] == 'automobile']
nombre_auto = prets_auto.shape[0]
print(nombre_auto)
cout_moyen_auto = prets_auto['cout_total'].mean()
print(cout_moyen_auto )
```

10. Afficher le bénéfice mensuel total réalisé par l’agence Parisienne. 

```python
benefice_paris = df[df['ville'] == 'PARIS']['benefices'].sum()
print(benefice_paris)
```
